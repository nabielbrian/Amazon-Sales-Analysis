### **1. Bagaimana distribusi penjualan berdasarkan kategori produk?**

- **Tujuan:** Mengidentifikasi produk yang paling menguntungkan dan segmen subkategori yang memberikan kontribusi terbesar terhadap revenue.

- **Insight:** 
  - **Grafik pertama (Bar chart kategori produk):** Menunjukkan distribusi total revenue berdasarkan kategori produk. Ini membantu untuk memahami kategori produk mana yang berkontribusi terbesar terhadap penjualan.

Ini penting untuk memprioritaskan produk unggulan dan mengoptimalkan strategi pemasaran.

- **Metode yang Digunakan:** 
  - **Grafik pertama:** Bar chart untuk total revenue (sumbu Y) yang dikelompokkan berdasarkan kategori produk (sumbu X).
  - **Grafik kedua:** Bar chart untuk total revenue (sumbu Y) yang dikelompokkan berdasarkan kategori produk dan subkategori produk (sumbu X).

```python
# Mengelompokkan data berdasarkan kategori produk dan menghitung total pendapatan untuk setiap kategori
category_revenue = dataframe_cleaned[['Product Category', 'Revenue']].groupby('Product Category')['Revenue'].sum().reset_index()

# Mengurutkan kategori produk berdasarkan total pendapatan (Revenue) dari yang tertinggi
category_revenue = category_revenue.sort_values('Revenue', ascending=False)

# Membuat grafik batang untuk menampilkan distribusi pendapatan antar kategori produk
plt.figure(figsize=(10,6))
plt.bar(category_revenue['Product Category'], category_revenue['Revenue'], color='skyblue')
plt.xlabel('Kategori Produk') # Memberikan label sumbu X
plt.ylabel('Total Pendapatan Amazon')
plt.title('Distribusi Pendapatan Amazon Berdasarkan Kategori Produk')
plt.tight_layout()  # Menghindari pemotongan label
plt.show()
```

### **4. Apa tren penjualan keseluruhan produk per bulan?**

Tujuan: Mengidentifikasi pola musiman dalam penjualan dan mengetahui bulan-bulan dengan performa penjualan tertinggi atau terendah.

- **Insight:** Visualisasi ini menggunakan **line chart** untuk menunjukkan tren penjualan bulanan. Dengan melihat data ini, kita dapat mengidentifikasi bulan-bulan dengan penjualan tertinggi dan terendah, serta mendeteksi pola musiman atau fluktuasi yang terjadi setiap tahun. Hal ini akan membantu Amazon dalam merencanakan strategi penjualan, promosi, dan inventarisasi untuk bulan-bulan tertentu yang lebih menguntungkan.

- **Metode yang Digunakan:**  
  - **Line Chart**  untuk menggambarkan total revenue (sumbu Y) per bulan (sumbu X) yang menunjukkan fluktuasi dan tren penjualan sepanjang waktu.


```python
# Ekstrak Tahun dan Bulan
dataframe_cleaned['Year'] = dataframe_cleaned['Date'].dt.year
dataframe_cleaned['Month'] = dataframe_cleaned['Date'].dt.month

# Menghitung total revenue per bulan dan tahun
monthly_revenue = dataframe_cleaned.groupby(['Year', 'Month'])['Revenue'].sum().reset_index()

# Membuat plot line chart dengan penyesuaian sumbu X agar lebih rapi
plt.figure(figsize=(14, 8))

# Menggunakan 'Year-Month' sebagai sumbu X agar terlihat lebih jelas
x_labels = monthly_revenue['Year'].astype(str) + '-' + monthly_revenue['Month'].astype(str)

plt.plot(x_labels, monthly_revenue['Revenue'], color="Slateblue", alpha=0.8, linewidth=2)

# Menambahkan label dan judul
plt.title('Tren Penjualan Bulanan', fontsize=16)
plt.xlabel('Bulan-Tahun', fontsize=12)
plt.ylabel('Total Revenue Amazon(USD)', fontsize=12)

# Menambahkan label untuk bulan-tahun
plt.xticks(rotation=45, fontsize=10)

# Menampilkan plot
plt.tight_layout()
plt.show()
```

### **5. Bagaimana statistik deskriptif dari total revenue, unit cost, dan quantity yang terjual, dan apakah terdapat nilai ekstrim (outlier) dalam data?**


- **Tujuan:** Menyediakan gambaran umum distribusi data untuk total revenue, unit cost, dan quantity yang terjual, serta mengidentifikasi apakah terdapat outlier dalam data.
  
- **Insight:** Menggunakan statistik deskriptif untuk memahami nilai rata-rata, penyebaran data, dan pola distribusi. Analisis outlier juga dilakukan untuk mengidentifikasi data yang jauh dari nilai normal.

- **Metode yang Digunakan:**  
  - **Central Tendency**: Menggunakan **mean**, **median** untuk mengukur nilai pusat data.
  - **Measure of Variance**: Menggunakan **standard deviation**, **IQR** untuk mengukur sebaran data.
  - **Outlier Analysis**: Menggunakan **IQR** dan batas **lower limit** dan **upper limit** untuk mendeteksi nilai ekstrim dalam data.

```python
numerical_columns = ['Quantity', 'Unit Cost', 'Unit Price', 'Revenue']  

for col in numerical_columns:
    mean_value = dataframe_cleaned[col].mean()
    median_value = dataframe_cleaned[col].median()
    std_dev = dataframe_cleaned[col].std()
    skewness = dataframe_cleaned[col].skew()
    kurtosis = dataframe_cleaned[col].kurt()

    print(f'Statistik Deskriptif untuk {col}:')
    print(f'Mean (Rata-rata): {mean_value}')
    print(f'Median (Tengah): {median_value}')
    print(f'Standard Deviation (Deviasi Standar): {std_dev}')
    print(f'Skewness (Kemiringan): {skewness}')
    print(f'Kurtosis (Kurtosis): {kurtosis}')
    print('-' * 50) #buat bikin garis garis
```

```python
numerical_columns = ['Quantity', 'Unit Cost', 'Unit Price', 'Revenue']  

for col in numerical_columns:
    Q1 = dataframe_cleaned[col].quantile(0.25)
    Q3 = dataframe_cleaned[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR
    
    # Menghitung jumlah outlier
    outliers = dataframe_cleaned[(dataframe_cleaned[col] < lower_limit) | (dataframe_cleaned[col] > upper_limit)]
    num_outliers = outliers.shape[0]
    total_data = dataframe.shape[0]
    
    # Menghitung persentase outlier
    outlier_percentage = (num_outliers / total_data) * 100
    
    # Menampilkan hasil
    print(f'Statistik Deskriptif untuk {col}:')
    print(f'Lower Limit: {lower_limit}')
    print(f'Upper Limit: {upper_limit}')
    print(f'Jumlah Outliers: {num_outliers}')
    print(f'Persentase Outliers: {outlier_percentage:.2f}%')
    print('-' * 50)
```

### **6. Apakah terdapat perbedaan yang signifikan dalam rata-rata pendapatan (Revenue) antara pelanggan perempuan dan laki-laki di Amazon?**


- Tujuan:
  Menguji perbedaan yang signifikan dalam **rata-rata pendapatan Amazon** antara **pelanggan perempuan** dan **laki-laki**.

- Insight:
  Menggunakan **Mann-Whitney U Test** untuk menguji apakah ada perbedaan signifikan antara pendapatan yang diperoleh dari pelanggan perempuan dan laki-laki, mengingat data tidak terdistribusi normal.

- Metode yang Digunakan:
  - **Mann-Whitney U Test** digunakan karena data tidak memenuhi asumsi normalitas, untuk menguji perbedaan antara dua grup independen.

- Hipotesis:
  - **H0**: Tidak ada perbedaan rata-rata pendapatan Amazon untuk pelanggan perempuan dan laki-laki.
  - **H1**: Ada perbedaan rata-rata pendapatan Amazon antara pelanggan perempuan dan laki-laki.

- Kesimpulan:
  Berdasarkan **p-value**, jika **p-value < 0.05**, kita menolak **H0** dan menerima **H1**, yang berarti ada perbedaan yang signifikan antara pendapatan pelanggan perempuan dan laki-laki.

```python
# Pisahkan data berdasarkan jenis kelamin
female_data = dataframe_cleaned[dataframe_cleaned['Customer Gender'] == 'F']['Revenue']
male_data = dataframe_cleaned[dataframe_cleaned['Customer Gender'] == 'M']['Revenue']

# Uji Mann-Whitney U Test untuk perbedaan pendapatan antara perempuan dan laki-laki
u_stat, p_value = stats.mannwhitneyu(female_data, male_data)

# Menampilkan hasil uji
print(f'Mann-Whitney U Test: U-statistic = {u_stat:.2f}, p-value = {p_value:.4f}')

# Kesimpulan berdasarkan p-value
if p_value < 0.05:
    print('Tolak H0: Terima H1, ada perbedaan rata-rata pendapatan Amazon antara pelanggan perempuan dan laki-laki.')
else:
    print('Gagal menolak H0: Terima H0, tidak ada perbedaan rata-rata pendapatan Amazon antara pelanggan perempuan dan laki-laki.')
```

##### **Analisis Hasil Uji Mann-Whitney U Test:**

- **Tidak Ada Perbedaan Signifikan**:  
  Berdasarkan hasil uji Mann-Whitney U Test dengan p-value sebesar 0.2144, yang lebih besar dari tingkat signifikansi yang biasa digunakan (misalnya 0.05), kita gagal menolak hipotesis nol (H0). Ini berarti tidak ada perbedaan yang signifikan antara rata-rata pendapatan Amazon untuk pelanggan perempuan dan laki-laki.
- Dengan demikian, strategi pemasaran dan penjualan Amazon tidak perlu membedakan antara jenis kelamin dalam pendekatan mereka, karena hasil ini menunjukkan bahwa pendapatan tidak dipengaruhi oleh jenis kelamin.
